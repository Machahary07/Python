```python
class Parent():
    def show(self):
        print("Inherited from Parent class")

class Child(Parent):
    def display(self):
        print("Child class")

obj = Child()
obj.show()
obj.display()
```

```python
class Student():
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks
    
    def display_details(self):
        print(self.name, self.roll_no, self.marks)

    def calculate_total(self):
        return sum(self.marks)
    
    def calculate_percentage(self):
        return self.calculate_total()/3
    
    def grade(self):
        p = self.calculate_percentage()
        if p >= 90:
            return "A"
        elif p >= 75:
            return "B"
        elif p >= 50:
            return "C"
        else:
            return "D"
        
s1 = Student("A", 1, [80,90,100])
s2 = Student("B", 2, [40, 45, 50])

for s in (s1,s2):
    s.display_details()
    print("Total",s.calculate_total(),"Percentage",s.calculate_percentage(),"Grade",s.grade())
    print("")
```

```python
class Academic:
    def __init__(self, marks):
        self.marks = marks

    def calculate_total(self):
        return sum(self.marks)

    def calculate_percentage(self):
        return self.calculate_total() / len(self.marks)

class Personal:
    def __init__(self, name, roll_no):
        self.name = name
        self.roll_no = roll_no

    def display_details(self):
        print("Name:", self.name)
        print("Roll No:", self.roll_no)

class Student(Academic, Personal):
    def __init__(self, name, roll_no, marks):
        Personal.__init__(self, name, roll_no)
        Academic.__init__(self, marks)

    def grade(self):
        p = self.calculate_percentage()
        if p >= 90:
            return "A"
        elif p >= 75:
            return "B"
        elif p >= 50:
            return "C"
        else:
            return "Fail"

# Creating objects
s1 = Student("Amit", 101, [85, 90, 80])
s2 = Student("Riya", 102, [60, 70, 65])

# Displaying details
for s in (s1, s2):
    s.display_details()
    print("Total Marks:", s.calculate_total())
    print("Percentage:", s.calculate_percentage())
    print("Grade:", s.grade())
    print()
```

```python
class Academic():
    def __init__(self, marks):
        self.marks = marks

    def total_marks(self):
        return sum(self.marks)

    def percentage_marks(self):
        return self.total_marks()/len(self.marks)
    
class Personal():
    def __init__(self, name, roll):
        self.name = name
        self.roll = roll

    def display_details(self):
        print("Name: ",self.name)
        print("Roll_no:",self.roll)

class Student(Academic, Personal):
    def __init__(self,name,roll,marks):
        Personal.__init__(self, name, roll)
        Academic.__init__(self, marks)

    def grade(self):
        p = self.percentage_marks()
        if p >= 90:
            return "A"
        elif p >= 75:
            return "B"
        elif p >= 50:
            return "C"
        else:
            return "Fail"

# Creating objects
s1 = Student("Amit", 101, [85, 90, 80])
s2 = Student("Riya", 102, [60, 70, 65])

# Displaying details
for s in (s1, s2):
    s.display_details()
    print("Total Marks:", s.total_marks())
    print("Percentage:", s.percentage_marks())
    print("Grade:", s.grade())
    print()
```

```python
#overloading
class Number:
    def __init__(self, x):
        self.x = x

    def __add__(self, other):
        return self.x + other.x

n1 = Number(10)
n2 = Number(20)
print(n1 + n2)
```

```python
# Creating a list
numbers_list = [10, 20, 30, 40]

# Creating a tuple
numbers_tuple = (50, 60, 70)

# List operations
numbers_list.append(50)
numbers_list.extend([60, 70])
numbers_list[1] = 25    # modification allowed

# Tuple operations
print("Tuple element:", numbers_tuple[1])
print("Tuple slice:", numbers_tuple[0:2])
print("Tuple concatenation:", numbers_tuple + (80, 90))

# Display list and tuple
print("List:", numbers_list)
print("Tuple:", numbers_tuple)
```

```python
# Tuple is an ordered immutable collection of elements
# Tuple creation
t1 = (10, 20, 30)
t2 = (40, 50)

# Tuple operations
print(t1, t2)
print("Element at index 1:", t1[1]) # Access an element
print("Slice:", t1[0:2])            # Extracts a part of the tuple
print("Concatenation:", t1 + t2)    # Joins two or more tuples
print("Membership:", 20 in t1)      # Check whether the element is there in the tuple

# Function returning multiple values using tuple
def calculate(a, b):
    return a + b, a - b, a * b

result = calculate(10, 5)
print("Multiple return values:", result)

# Add new elements to the list
lst = list(t1)
lst.append(5) #type: ignore
t1 = tuple(lst)

print("Modified tuple", t1)
```

```python
# Original tuple
t = (10, 20, 30)

# Convert tuple to list
lst = list(t)

# Modify the list
lst[1] = 25 #type: ignore #put new element 25 at index 1
lst.append(40) #type: ignore #push element 40 at the end

# Convert back to tuple
t = tuple(lst)

print("Modified tuple:", t)
```

```python
def check_number(n, limit=5):  # Function with default argument
    for i in range(1, limit + 1):      # Loop statement
        if i == 3:
            continue                   # Skips iteration when i = 3
        if i > n:
            break                      # Exits loop
        pass                           # Does nothing (placeholder)
        print(i)

    if n < 0:                          # Conditional statement
        print("Negative number")
        exit()                         # Terminates program
    else:
        print("Positive number")

num = int(input("Enter a number: "))   # Input statement
print("You entered:", num)             # Output statement

check_number(num)                      # Function call
```

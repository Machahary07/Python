```python
books = {
  1: {"Author": "Neelashree", "Name": "Life", "Price": "₹400"},
  2: {"Author": "Jeu", "Name": "Code", "Price": "₹300"}
}

for book_id, details in books.items():
    print(f"Book ID: {book_id}")
    print(f"Author : {details['Author']}")
    print(f"Name   : {details['Name']}")
    print(f"Price  : {details['Price']}\n")
```

```python
# Example of a Variable, Identifier, and Literal

# 1️⃣ Variables
age = 15              # "age" is the variable (box), 15 is the value inside
name = "Alice"        # "name" is a variable storing the string "Alice"
is_student = True     # Boolean variable storing True

# 2️⃣ Identifiers
# The names: age, name, is_student → are identifiers
# They follow Python naming rules (no starting with a number, no keywords)

# 3️⃣ Literals
# The values: 15, "Alice", True are literals
# These are fixed values written directly in the code

# Printing variables
print("Age:", age)
print("Name:", name)
print("Is student?", is_student)

# Changing a variable's value
age = 16   # We replaced the old value (15) with the new one (16)
print("Updated Age:", age)

# Collection literal examples
fruits = ["apple", "banana", "cherry"]  # list literal
details = {"city": "London", "zip": 12345}  # dictionary literal

print("Fruits:", fruits)
print("Details:", details)
```

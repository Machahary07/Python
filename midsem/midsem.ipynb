```python
# WAF that takes length and breadth of a rectangle & returns its perimeter and area
def rect(l, b):
    a = l * b
    p = 2 * (l + b)
    return a, p

a, p = rect(2, 3)
print("Area:", a, "Perimeter:", p)
```

```python
#WAF that takes radius of a circle & returns its circumference and area without import math
def circle(r):
    pi = 3.14
    a = pi * r * r
    c = 2 * pi * r
    return a, c 

a, c = circle(7)
print("Area:", a, "Circumference:", c)
```

```python
# WAF to find the nth fibonacci number using recursion
def fib(n):
    if n <= 1:
        return n
    else:
        return fib(n-1) + fib(n-2)

print(fib(6))
```

```python
# WAF that takes a list of numbers and returns two lists - one containing odd numbers and the other containing even numbers
def odd_even(lst):
    odd = []
    even = []
    for i in lst:
        if i % 2 == 0:
            even.append(i)
        else:
            odd.append(i)
    return odd, even

odd, even = odd_even([1, 2, 3, 4, 5, 6])
print("Odd:", odd, "Even:", even)
```

```python
#WAF to count vowels in a given string
def count_vowels(s):
    vowels = 'aeiouAEIOU'
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

print(count_vowels("Hello, World!"))
```

```python
#WAF to check if a string is a palindrome
def is_pal(s: str)-> bool:
    return s == s[::-1]

def is_pal_int(n: int) -> bool:
    o, rev = n, 0
    while n > 0:
        rev = rev * 10 + n % 10
        n //= 10
    return o == rev

print(is_pal("woo"))
print(is_pal("madam"))
print(is_pal("1234321"))
print(is_pal("1234"))

print(is_pal_int(123))
print(is_pal_int(12321))
```

```python
#WAF that accepts a sentence and returns the word with the maximum length
def max_l_word(s: str) -> str:
    return max(s.split(), key=len)

print(max_l_word("Hello World from Python"))
```

```python
#WAF that acceps a list of numbers and returns the second largest number
def second_largest(lst: list) -> int:
    if len(lst) < 2:
        raise ValueError("List must contain at least two elements")
    unique_lst = list(set(lst))  # Remove duplicates
    unique_lst.sort()
    return unique_lst[-2]  # Second largest
print(second_largest([1, 2, 3, 4, 5, 5, 4, 3, 2, 1]))
```

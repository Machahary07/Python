```python
#Q5) Write a program to display the following figure (there are spaces at the start)
#     *
#    * *
#   * * *
#  * * * *
# * * * * *

for i in range(5):
    for j in range(5-i):
        print(" ",end=" ")
    for k in range(i+1):
        print("*",end=" ")
    print()
```

```python
#Q8) Write a program to print first 10 multiples of a number.

def mn(n:int):
    for i in range(n<10):
        print(i*i)
    
mn(2)
```
